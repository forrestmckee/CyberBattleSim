```python
import gym
import numpy as np
import cyberbattle
```

Make two copies of the same environment. Discover NodeA first on `gym_env_a`. It has index 1 internally. Discover NodeB first on `gym_env_b`. It has index 1 internally.

```python
gym_env_a = gym.make("CyberBattleInternalIDError-v0")
gym_env_a.reset()
```

```python
print(gym_env_a.env._CyberBattleEnv__discovered_nodes)
print(gym_env_a.env._CyberBattleEnv__find_external_index('foothold'))
print(gym_env_a.env._CyberBattleEnv__internal_node_id_from_external_node_index(0))
```

```python
vuln_a = {'local_vulnerability': np.array([0, 0])}
gym_env_a.step(vuln_a)
```

```python
print(gym_env_a.env._CyberBattleEnv__discovered_nodes)
print(gym_env_a.env._CyberBattleEnv__find_external_index('foothold'))
print(gym_env_a.env._CyberBattleEnv__internal_node_id_from_external_node_index(0))
print(gym_env_a.env._CyberBattleEnv__find_external_index('NodeA'))
print(gym_env_a.env._CyberBattleEnv__internal_node_id_from_external_node_index(1))
```

```python
gym_env_b = gym.make("CyberBattleInternalIDError-v0")
gym_env_b.reset()
```

```python
print(gym_env_b.env._CyberBattleEnv__discovered_nodes)
print(gym_env_b.env._CyberBattleEnv__find_external_index('foothold'))
print(gym_env_b.env._CyberBattleEnv__internal_node_id_from_external_node_index(0))
```

```python
vuln_b = {'local_vulnerability': np.array([0, 1])}
gym_env_b.step(vuln_b)
```

```python
print(gym_env_b.env._CyberBattleEnv__discovered_nodes)
print(gym_env_b.env._CyberBattleEnv__find_external_index('foothold'))
print(gym_env_b.env._CyberBattleEnv__internal_node_id_from_external_node_index(0))
print(gym_env_b.env._CyberBattleEnv__find_external_index('NodeB'))
print(gym_env_b.env._CyberBattleEnv__internal_node_id_from_external_node_index(1))
```

`networkx` nodes are equivalent, but interal id is dependent on the order of vulnerability execution.

```python
gym_env_a.env.environment.network.nodes,gym_env_b.env.environment.network.nodes
```

Action masks are the same although different vulnerabilities are executed and different nodes are discovered.

```python
mask_a=gym_env_a.compute_action_mask()
mask_b=gym_env_b.compute_action_mask()
```

```python
all(np.array_equal(mask_a[key],mask_b[key]) for key in mask_a)
```
